## FINDING AND HANDLING THE MISSING DATA

```python
netflix_df.isnull().sum()
```

```python
# The below code helps to calculate the percentage of missing values in each column

(netflix_df.isnull().sum()/(len(netflix_df)))*100
```

Since the columns 'director','cast','country' have maximum number of missing values we are going to handle the missing values in those columns.

```python
netflix_df[netflix_df['director'].isnull()]
```

```python
#filling the NaN in director column with unknown.
netflix_df['director'] = netflix_df['director'].fillna('unknown')
netflix_df['director'].isnull().sum()
```

```python
netflix_df[netflix_df['cast'].isnull()]
```

```python
#replacing nan values in cast with unknown
netflix_df['cast'] = netflix_df['cast'].fillna('unknown')
netflix_df['cast'].isnull().sum()
```

```python
netflix_df[netflix_df['country'].isnull()]
```

```python
#filling the NaN in country column with unknown.
netflix_df['country'] = netflix_df['country'].fillna('Unknown')
netflix_df['country'].isnull().sum()
```

```python
#dropping rows with missing values

netflix_df.dropna(inplace=True)
netflix_df.isnull().sum()
```

```python
netflix_df[['date_added','release_year']]
```

```python
#we are splitting the date added into month added and year added
netflix_df['added_month'] = netflix_df['date_added'].apply(lambda x:x.split(',')[0].split()[0])
netflix_df['added_year'] = netflix_df['date_added'].apply(lambda x:x.split(',')[1])
netflix_df[['added_month','added_year']]
```

```python
# we are dropping date added colum and adding the added year
netflix_df.drop('date_added', axis=1, inplace = True)
netflix_df['added_year'] = netflix_df['added_year'].astype(int)
```

```python
#Droping rows in which added_year is less than release_year¶
netflix_df.drop(netflix_df[netflix_df['added_year']<netflix_df['release_year']].index, inplace=True)
```

```python
#checking the type of our features.

numeric_features = [feature for feature in netflix_df.columns if netflix_df[feature].dtype != 'O']
categorical_features = [feature for feature in netflix_df.columns if netflix_df[feature].dtype == 'O']

# print columns
print('We have {} numerical features : {}'.format(len(numeric_features), numeric_features))
print('\nWe have {} categorical features : {}'.format(len(categorical_features), categorical_features))
```
